CONDICIONAIS

```python
### if e else ###

idade = int(input("Digite a sua idade: "))

if idade >= 18:
  print("Você é maior de idade")
else:
  print("Você é menor de idade")
```

```python
idade = int(input("Digite a sua idade: "))
if idade < 16:
  print("Você não pode votar")
elif idade >= 16 and idade < 18:
  print("Voto opicional")
else:
  print("Voto obrigatorio")
```

ATIVIDADES


```python
### ATIVIDADE 01 ###
# A ORDEM DO elif IMPORTA!! #
pagamento = input("Digite o tipo de pagamento: ")
if pagamento == "Cartão":
  print("Processando pagamento no cartão")
elif pagamento == "Boleto":
  print("Gerando pagamento pelo boleto")
else:
  print("Metado de de pagamento não reconhecido")
```

```python
### ATIVIDADE 02 ###
idade = int(input("Digite a sua idade: "))
if idade < 12:
  print("Criança")
elif idade >= 13 and idade <= 17:
  print("Adolescente")
elif idade >= 18 and idade <= 59:
  print("Adulto")
else:
  print("Idoso")
```

```python
### ATIVIDADE 04 ###
entrega = input("Metado de entrega escolhido (Padrão/Expresso): ")
if entrega == "Padrão":
  print("Valor custiado: R$ 10,00")
elif entrega == "Expresso":
  print("Valor custiado: R$ 20,00")
else:
  print("Opção de entrega inválida")
```
